```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Rhayven\Downloads\customer_feedback_satisfaction.csv')
```

```python
df.head()
```

```python
df.describe()

df.isnull().sum()
```

```python
bins = [18, 30, 50, 70, 90]
labels = ['18-30', '31-50', '51-70', '71-90']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='AgeGroup', palette='Set2')
plt.title('Customer Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', palette='Set3')
plt.title('Customer Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Country', palette='Set1')
plt.title('Customer Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='LoyaltyLevel', palette='Set3')
plt.title('Customer Distribution by Loyalty Level')
plt.xlabel('Loyalty Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='SatisfactionScore')
plt.title('Age vs. Satisfaction Score')
plt.xlabel('Age')
plt.ylabel('Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Income', bins=20, kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='SatisfactionScore')
plt.title('Satisfaction Score by Country')
plt.xlabel('Country')
plt.ylabel('Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='FeedbackScore', y='SatisfactionScore')
plt.title('Feedback Score vs. Satisfaction Score')
plt.xlabel('Feedback Score')
plt.ylabel('Satisfaction Score')
plt.show()
```
